### Import

````python
import numpy as np
import sympy as sp
from sympy.plotting import plot3d
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse

x, y = sp.symbols('x y')

def print_(symfunc, x, y, start, stop) -> None:
    f = sp.lambdify(sp.Symbol('x'), symfunc)
    actual = [f(x_) for x_ in x]
    x_range = np.linspace(start, stop, num=100)
    plt.plot(x_range, f(x_range))
    plt.scatter(x, y, color='red')
    plt.show()
    print(f'MSE = {mse(y, actual, squared=True)}')
````

## Раздел 1
### 1. Получите значение MSE меньшее 0.75.

````python
points_x = [-2, -1, 0, 1, 2, 3, 3.5, 4, 4.5, 5]
points_y = [3, 1, -0.5, 1.2, 2.5, 0.8, -2, -3.5, -3, -5]
f = 2 * x * sp.sin(2 * x + 5) + 5 * sp.sin(2 * x + 5) + 2
f_new = f.subs(x, (x-5)/2) * 1.2 - 2
````

````python
print_(f_new, points_x, points_y, -10, 10)
````
