## Нули функций нескольких переменных. Практика
Для следующих функций найдите решения и постройте их графики, используя sympy:
1. $f(x, y) = 2x + 5y + 1$
2. $f(x, y) = x^2 + y^2 - 1$
3. $f(x, y) = xcos(y)$

### Import

```python
import sympy as sp
from sympy.plotting import plot3d

x, y = sp.symbols('x y')
```

### Solve 1

```python
f = 2*x + 5*y + 1
sp.solve(f)
```

### Plot 1

```python
plot3d(f)
```

### Solve 2

```python
f = x**2 + y**2 - 1
sp.solve(f)
```

### Plot 2

```python
plot3d(f)
```

### Solve 3

```python
f = x*sp.cos(y)
sp.solve(f)
```

### Plot 3

```python
plot3d(f)
```
