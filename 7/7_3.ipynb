### Import

```python
import sympy as sp
from sympy.plotting import plot3d
x, y = sp.symbols('x y')
```

### Plot 1

```python
f = sp.sin(sp.sqrt(x**2 + y**2))
plot3d(f)
```

### Plot 2

```python
f = sp.exp(-0.9*x**2-0.45*(x-y)**2)
plot3d(f, (x, -3, 3), (y, -3, 3))
```

### Plot 3

```python
f = x*sp.cos(y)
plot3d(f)
```
