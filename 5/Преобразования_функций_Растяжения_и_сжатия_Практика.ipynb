### Import

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse

%matplotlib inline
```

### Create function for printing plot with points and RMSE

```python
def print_(symfunc, x, y, start, stop) -> None:
    f = sp.lambdify(sp.Symbol('x'), symfunc)
    actual = [f(x_) for x_ in x]
    x_range = np.linspace(start, stop, num=100)
    plt.plot(x_range, f(x_range))
    plt.scatter(x, y, color='red')
    plt.show()
    print(f'RMSE = {mse(y, actual, squared=True)}')
```

### 1. Получите значение RMSE меньшее 50.

```python
points_x = [-2, -1, 0, 1, 2, 3, 3.5, 4, 4.5, 5]
points_y = [-15, -1, 4, -9, -2, -5, -8, 4, 13, 21]
x = sp.Symbol('x')
f1 = 3.375*x**3 - 9*x**2
f_new = f1.subs(x, x / 1.5) / 1.5
```

```python
print_(f_new, points_x, points_y, -2, 5)
```

### 2. Получите значение RMSE меньшее 150.

```python
points_x = [-3, -2, -1, 1, 2, 3]
points_y = [-55, -40, 7, 5, 38, 53]
x = sp.Symbol('x')
f2 = 0.074*x**3 - 0.11*x**2 + x + 5
f_new = f2.subs(x, (x - 0.2) * 2) * 2.6
```

```python
print_(f_new, points_x, points_y, -3, 3)
```
