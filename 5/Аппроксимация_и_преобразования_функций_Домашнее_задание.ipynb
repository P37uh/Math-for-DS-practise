### Import

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

%matplotlib inline
```

### Solve 1

```python
a0, a1, a2, a3, a4, a5 = sp.symbols('a:6')
eq1 = a5*(-4)**5 + a4*(-4)**4 + a3*(-4)**3 + a2*(-4)**2 + a1*(-4) + a0 + 4268
eq2 = a5*(-3)**5 + a4*(-3)**4 + a3*(-3)**3 + a2*(-3)**2 + a1*(-3) + a0 + 1227
eq3 = a5*(-1)**5 + a4*(-1)**4 + a3*(-1)**3 + a2*(-1)**2 + a1*(-1) + a0 + 17
eq4 = a5 + a4 + a3 + a2 + a1 + a0 - 17
eq5 = a5*3**5 + a4*3**4 + a3*3**3 + a2*3**2 +a1*3 + a0 - 1227
eq6 = a5*4**5 + a4*4**4 + a3*4**3 + a2*4**2 + a1*4 + a0 - 4268
sp.nonlinsolve([eq1, eq2, eq3, eq4, eq5, eq6], sp.symbols('a:6'))
```

### Plot 1

```python
x_range = np.linspace(-5, 5, num=100)
func = lambda x: 3*x**5 + 19*x**3 - 5*x
plt.axhline(0, 0, color='black')
plt.axvline(0, 0, color='black')
plt.plot(x_range, func(x_range))
plt.scatter([-4, -3, -1, 1, 3, 4], [-4268, -1227, -17, 17, 1227, 4268], color='red')
```

### Import

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse

%matplotlib inline
```

### Create function for printing plot with points and RMSE

```python
def print_(symfunc, x, y, start, stop) -> None:
    f = sp.lambdify(sp.Symbol('x'), symfunc)
    actual = [f(x_) for x_ in x]
    x_range = np.linspace(start, stop, num=100)
    plt.plot(x_range, f(x_range))
    plt.scatter(x, y, color='red')
    plt.show()
    print(f'RMSE = {mse(y, actual, squared=True)}')
```

### 1. Получите значение RMSE меньшее 5.

```python
points_x = [-2, -1, 0, 1, 2, 3, 3.5, 4, 4.5, 5]
points_y = [2, -4, 1, 8, 21, 40, 47, 65, 75, 92]
x = sp.Symbol('x')
f = 48*x**2 - 128*x + 83
f_new = f.subs(x, x+2.3) / 18 - 3
```

```python
plt.ylim(-10, 100)
print_(f_new, points_x, points_y, -3, 6)
```

```python
plt.ylim(-40, 200)
print_(f_new, points_x, points_y, -3, 6)
```

```python
plt.ylim(-850, 100)
print_(f_new, points_x, points_y, -10, 6)
```
