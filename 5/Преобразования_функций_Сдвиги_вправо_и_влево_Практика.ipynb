### Import

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse

%matplotlib inline
```

### Create function for printing plot with points and RMSE

```python
def print_(symfunc, x, y, start, stop) -> None:
    f = sp.lambdify(sp.Symbol('x'), symfunc)
    actual = [f(x_) for x_ in x]
    x_range = np.linspace(start, stop, num=100)
    plt.plot(x_range, f(x_range))
    plt.scatter(x, y, color='red')
    plt.show()
    print(f'MSE = {mse(y, actual, squared=True)}')
```

### 1. Получите значение MSE меньшее 110.

```python
points_x = [-3, -2, -1, 1, 2, 3]
points_y = [60, 30, 19, -16, -37, -23]
x = sp.Symbol('x')
f1 = -14*x - 100
f_new = f1.subs(x, x-7.5)
```

```python
print_(f_new, points_x, points_y, -3, 3)
```

### 2. Получите значение MSE меньшее 150.

```python
points_x = [-3, -2, -1, 1, 2, 3]
points_y = [-55, -40, 7, 5, 38, 53]
f2 = 2*x**3 - 121*x**2 + 2443*x - 16455
f_new = f2.subs(x, x + 20)
```

```python
print_(f_new, points_x, points_y, -3, 3)
```
