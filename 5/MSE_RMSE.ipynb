## Практическое задание.
Для того, чтобы закрепить материал, посчитайте значение среднеквадратичной
ошибки для данных наборов точек и аппроксимирующих функций. Постройте графики
этих функций и отметьте данные точки на нем, чтобы посмотреть, как проходит через них
график аппроксимирующей функции. При решении можете пользоваться python и
matplotlib, а можете по-старинке, сделать это на бумаге, хотя это будет не слишком
удобно.

Начальный уровень.
1. $(2; 12.258), (4; 17.24), (8; 30.151).$
- $f(x) = 3x + 5$
2. $(2; 3.688), (4; 10.791), (8; 20.705).$
- $f(x) = 0.25x^2 + 0.75x + 1.25$

Средний уровень.
1. $(2; 4.872), (4; 29.707), (8; 246.971), (10; 485.727), (12; 840.658).$
- $f(x) = 0.5x^3 - 0.25x^2 + 0.75x + 1.25$

### Import

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse

'''
Just for understanding:

def mse(actual: np.ndarray, pred: np.ndarray) -> float:
    return np.square(np.subtract(actual,pred)).mean()
    
def rmse(actual: np.ndarray, pred: np.ndarray) -> float:
    return np.sqrt(mse(actual, pred))
'''
    
```

### Solve 1

```python
actual = [12.258, 17.24, 30.151]
x = sp.Symbol('x')
f = lambda x: 3*x + 5
pred = [f(2), f(4), f(8)]
mse(actual, pred, squared=True)
```

### Plot 1

```python
x_range = np.linspace(0, 10, num=2)
plt.plot(x_range, f(x_range))
plt.scatter([2, 4, 8], actual, color='red')
```

### Solve 2

```python
actual = [3.688, 10.791, 20.705]
x = sp.Symbol('x')
f = lambda x: 0.25*x**2 + 0.75*x + 1.25
pred = [f(2), f(4), f(8)]
mse(actual, pred, squared=True)
```

### Plot 2

```python
x_range = np.linspace(0, 10, num=100)
plt.plot(x_range, f(x_range))
plt.scatter([2, 4, 8], actual, color='red')
```

### Solve 3

```python
actual = [4.872, 29.707, 246.971, 485.727, 840.658]
x = sp.Symbol('x')
f = lambda x: 0.5*x**3 - 0.25*x**2 + 0.75*x + 1.25
pred = [f(2), f(4), f(8), f(10), f(12)]
mse(actual, pred, squared=True)
```

### Plot 3

```python
x_range = np.linspace(0, 14, num=100)
plt.plot(x_range, f(x_range))
plt.scatter([2, 4, 8, 10, 12], actual, color='red')
```
