А можете решать вручную и затем проверять себя.

1. $f(x) = x^2 + 3x - 4, f'(x) = 2x + 3$
2. $f(x) = x^3 - 2x^2 + x - 6, f'(x) = 3x^2 - 4x + 1$
3. $f(x) = e^{2x-x^2}, f'(x) = (2-2x)e^{2x-x^2}$

### Import

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.Symbol('x')
```

### Solve 1

```python
f = x**2 + 3*x - 4
dydx = 2*x+3
sp.solve(dydx<0)
```

```python
sp.solve(dydx>0)
```

### Plot 1

```python
x_ = -(3/2)
y = f.subs(x, x_)
x_range = np.linspace(-7, 5, num=100)
plt.plot(x_range, sp.lambdify(x, f)(x_range))
plt.scatter(x_, y, color='red')
```

### Solve 2

```python
f = x**3 - 2*x**2 + x - 6
dydx = 3*x**2 - 4*x + 1
sp.solve(dydx<0)
```

```python
sp.solve(dydx>0)
```

### Plot 2

```python
x_ = [1, 1/3]
y = [f.subs(x, x_[0]), f.subs(x, x_[1])]
x_range = np.linspace(-0.5, 2, num=100)
plt.plot(x_range, sp.lambdify(x, f)(x_range))
plt.scatter(x_, y, color='red')
```

### Solve 3

```python
f = sp.exp(2*x-x**2)
dydx = (2-2*x)*sp.exp(2*x-x**2)
sp.solve(dydx<0)
```

```python
sp.solve(dydx>0)
```

### Plot 3

```python
x_ = 1
y = f.subs(x, x_)
x_range = np.linspace(-10, 10, num=100)
plt.plot(x_range, sp.lambdify(x, f)(x_range))
plt.scatter(x_, y, color='red')
```
